```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit
```

```python
def function (x, k):
    return k*pow(x,0.5);
```

```python
path = "Discrete_RW.out"

M, N, L = np.loadtxt(path, usecols=(0, 1, 2), delimiter=' ', unpack='true', max_rows=1)

d, err = np.loadtxt(path, usecols=(0, 1), delimiter=' ', unpack='true', skiprows=1)

x = np.arange(L+1)

plt.title("Discrete RW", color = 'bl')

plt.errorbar(x,d,yerr=err, label = 'simulated')
plt.xlabel('# steps')
plt.ylabel('mean value')
plt.grid(True)

pars, cov = curve_fit(f=function, xdata=x, ydata=d, p0=[1], bounds=(-np.inf, np.inf))
plt.plot(x, function(x, *pars), linestyle='--', linewidth=2, label = "fit")
plt.legend()

plt.show()
```
